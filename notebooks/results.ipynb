```python
import sys

import pandas as pd
import seaborn as sns

sys.path.append("..")

from utils import paths
from utils.data_visualization import plot_kfold_evaluation_results

sns.set_style("darkgrid")
pd.set_option('display.max_columns', None)
```

```python
cnn1 = pd.read_csv(f"{paths.RESULTS_PATH}/experiment_20230711_222446_results.csv")
cnn1["model"] = cnn1["model"].replace("experiment_20230711_222446", "Model 1")
cnn2 = pd.read_csv(f"{paths.RESULTS_PATH}/experiment_20230711_222532_results.csv")
cnn2["model"] = cnn2["model"].replace("experiment_20230711_222532", "Model 2")
cnn3 = pd.read_csv(f"{paths.RESULTS_PATH}/experiment_20230711_222628_results.csv")
cnn3["model"] = cnn3["model"].replace("experiment_20230711_222628", "Model 3")
cnn4 = pd.read_csv(f"{paths.RESULTS_PATH}/experiment_20230711_222914_results.csv")
cnn4["model"] = cnn4["model"].replace("experiment_20230711_222914", "Model 4")
cnn5 = pd.read_csv(f"{paths.RESULTS_PATH}/experiment_20230711_223049_results.csv")
cnn5["model"] = cnn5["model"].replace("experiment_20230711_223049", "Model 5")

kfold_results = pd.concat([cnn1, cnn2, cnn3, cnn4, cnn5])

kfold_results.sample(10)
```
