```python
import sys
import json

import pandas as pd

sys.path.append("..")

from utils import paths

pd.set_option("display.max_columns", None)
```

```python
data = pd.read_csv(paths.DATA_PATH + "/.230517COVID19MEXICO.csv", dtype="str")
print(f"Shape: {data.shape}")
data.sample(10)
```

# Scrub for irrelevant and duplicated data

```python
data = data.drop(columns=["FECHA_ACTUALIZACION", "ID_REGISTRO", "ENTIDAD_UM", "ENTIDAD_NAC", "ENTIDAD_RES", 
                          "MUNICIPIO_RES","HABLA_LENGUA_INDIG", "INDIGENA", "MIGRANTE", "NACIONALIDAD",
                          "ORIGEN", "SECTOR", "TOMA_MUESTRA_LAB", "RESULTADO_LAB", "TOMA_MUESTRA_ANTIGENO", 
                          "RESULTADO_ANTIGENO"])
data.sample(10)
```

# Fix dtypes

```python
data.dtypes
```

```python
numeric_columns = ["SEXO", "TIPO_PACIENTE", "INTUBADO", "NEUMONIA", "EDAD", "EMBARAZO", "DIABETES", "EPOC", "ASMA", 
                   "INMUSUPR", "HIPERTENSION", "OTRA_COM", "CARDIOVASCULAR", "OBESIDAD", "RENAL_CRONICA", "TABAQUISMO",
                   "OTRO_CASO", "CLASIFICACION_FINAL", "UCI"]
data[numeric_columns] = data[numeric_columns].astype("int8")
```

```python
date_columns = ["FECHA_INGRESO", "FECHA_SINTOMAS"]
data[date_columns] = data[date_columns].astype("datetime64[ns]")
```

```python
decision_column = "FECHA_DEF"
data[decision_column] = data[decision_column].apply(lambda x: 0 if x == "9999-99-99" else 1)
data[decision_column] = data[decision_column].astype("int8")
data = data.rename(columns={"FECHA_DEF": "DEFUNCION"})
```

```python
data["PAIS_NACIONALIDAD"] = data["PAIS_NACIONALIDAD"].astype("category")
PAIS_NACIONALIDAD_DICT = dict(zip(data["PAIS_NACIONALIDAD"].cat.codes, data["PAIS_NACIONALIDAD"]))
data["PAIS_NACIONALIDAD"] = data["PAIS_NACIONALIDAD"].cat.codes

with open(paths.DATA_PATH + "/PAIS_NACIONALIDAD.json", "w") as outfile:
    json.dump(PAIS_NACIONALIDAD_DICT, outfile, indent=4)

data["PAIS_ORIGEN"] = data["PAIS_ORIGEN"].astype("category")
PAIS_ORIGEN_DICT = dict(zip(data["PAIS_ORIGEN"].cat.codes, data["PAIS_ORIGEN"]))
data["PAIS_ORIGEN"] = data["PAIS_ORIGEN"].cat.codes

with open(paths.DATA_PATH + "/PAIS_ORIGEN.json", "w") as outfile:
    json.dump(PAIS_ORIGEN_DICT, outfile, indent=4)


```
